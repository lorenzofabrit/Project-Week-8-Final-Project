# Vehicles with low occupancy

### I will now try to find the vehicles with the lowest maximum occupancy (lower than 50%) during the two timeframes that I am analyzing. That way I will be able to to divert some of these vehicles and towards the overloaded lines, so as to better make use of the capacities of the Zurich public transport network. 

### Create a dictionary to find all the lines that never go above 50 % during the overused hours

```python
# lets start with buses
mydict_bus = {}
for line in lines:
    mydict_bus[line] = zh[(zh.line_name == line) & (zh.daytype == 0) & (zh.vehicle_type == 'bus')].loc[(zh.time > '07:00:00') & 
                        (zh.time < '09:00:00')].sort_values(by=['occupancy_rate']).occupancy_rate.max()
    
```

```python
mydict_bus
```

```python
# trams are next
mydict_tram = {}
for line in lines:
    mydict_tram[line] = zh[(zh.line_name == line) & (zh.daytype == 0) & (zh.vehicle_type == 'tram')].loc[(zh.time > '07:00:00') & 
                        (zh.time < '09:00:00')].sort_values(by=['occupancy_rate']).occupancy_rate.max()
```

```python
mydict_tram
```

```python
mydict_large_bus = {}
for line in lines:
    mydict_large_bus[line] = zh[(zh.line_name == line) & (zh.daytype == 0) & (zh.vehicle_type == 'large_bus')].loc[(zh.time > '07:00:00') & 
                        (zh.time < '09:00:00')].sort_values(by=['occupancy_rate']).occupancy_rate.max()
```
